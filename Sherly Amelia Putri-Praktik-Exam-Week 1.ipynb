```python
#Soal 1 
#Membuat fungsi faktorial

def faktorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * faktorial(n - 1)

#Faktorial bilangan
n = int(input('Masukkan bilangan bulat positif: '))
if n >= 0:
    print(f'Faktorial dari {n} adalah {faktorial(n)}')
else:
    print('Masukkan bilangan bulat positif!')
```

```python
#Soal 2 
#Menghitung jumlah huruf vokal

def hitung_vokal(kalimat):
    vokal = 'aeiou' #Merupakan variabel vokal
    hasil = {v: 0 for v in vokal}  #Inisialisasi kamus untuk vokal (v) memiliki awalan 0
    
    for huruf in kalimat.lower():  #Untuk mengubah kalimat menjadi huruf kecil
        if huruf in vokal:  #Jika huruf merupakan vokal
            hasil[huruf] += 1 #Maka hasil['huruf'] bertambah 1
    
    # Hanya mengembalikan vokal yang muncul (nilai lebih dari 0)
    hasil = {k: v for k, v in hasil.items() if v > 0}
    return hasil

# Input kalimat yang diinginkan
kalimat = input('Masukkan kalimat: ')
vokal_terhitung = hitung_vokal(kalimat)
print(f'Jumlah vokal dalam kalimat: {vokal_terhitung}')
```

```python
#Soal 3
#Membuat daftar angka genap dengan list comprehension

def angka_genap(daftar):
    return [x for x in daftar if x % 2 == 0] #List comprehension untuk angka genap

daftar_angka = list(range(1, 101)) #Membuat daftar angka 1-100
print(angka_genap(daftar_angka)) 
```

```python
#Soal 4
#Membuat program untuk memeriksa string termasuk palindrome (dibaca sama dari depan dan belakang)

def cek_palindrome(teks):
    #Mengubah kalimat menjadi huruf kecil 
    teks = teks.lower()
    #Memeriksa apakah kalimat sama jika dibalik
    return teks == teks[::-1]

input_teks = input('Masukkan kalimat: ')

#Memeriksa apakah kalimat adalah palindrome
if cek_palindrome(input_teks):
    print(f'{input_teks} adalah palindrome')
else:
    print(f'{input_teks} bukan palindrome')
```

```python
#Soal 5
#Mengurutkan daftar tuple berdasarkan elemen kedua
def urut(daftar_tuple):
    #Menggunakan sorted dengan key lambda untuk mengurutkan berdasarkan elemen kedua
    return sorted(daftar_tuple, key=lambda x: x[1])
    
daftar_tuple = [(2, 1), (1, 3), (7, 9), (6, 7)]
hasil = urut(daftar_tuple) #Fungsi untuk mengurutkan
print(hasil)
```

```python
#Soal 2
#Membuat matriks 4x4 dengan angka acak 

import numpy as np
matrix = np.random.randint(0, 11, (4, 4))
transpose_matrix = np.transpose(matrix) #Mencari transpose dari matriks
determinant = np.linalg.det(matrix) #Mencari determinan dari matriks

#Menghitung invers dari matriks jika determinan tidak nol
if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
else:
    inverse_matrix = 'Invers tidak dapat dihitung (determinannya nol)'

#Menampilkan hasil
print('Matriks Asli:\n', matrix)
print('Transpose Matriks:\n', transpose_matrix)
print('Determinan Matriks:', determinant)
print('Invers Matriks:\n', inverse_matrix)
```
