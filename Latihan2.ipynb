```python
#proses data cleaning
#menghapus kolom yang tidak dibutuhkan
#menghapus kolom row ID karna bukan data dengan perintah drop

df.drop('Row ID', axis = 1, inplace = True)

#axis = 1 artinya satu kolom penuh, inplace = true artinya hapus permanen
```

```python
#konversi tanggal order menjadi tanggal format pandas
df['Order Date']=pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

#convert shipping dates to pandas datetime format
df['Ship Date']=pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
#sorting data by order date sekaligus dibuat permanen
#urutan ini (inplace=true) dan secara ascending (kecil ke besar)

df.sort_values(by=['Order Date'], inplace=True, ascending=True)
```

```python
#menampilkan 20 orang customer yang paling banyak keluarkan uang di perusahaan kita

#group_by
Top_customers=df[['Customer Name','Sales']].groupby(['Customer Name']).sum()

#sort+values
#ascending
Top_customers = Top_customers.sort_values('Sales', ascending=False)

#head
Top_customers = Top_customers[['Sales']].round(2)

#memberi
Top_customers.reset_index(inplace=True)
Top_customers.head(10)
```

```python
import seaborn as sns

#mengcopy table df ke new df, sehingga step lanjutannya tidak mengganggu/merusak df
newdf=df.copy()

#step1: set 'Order Date' as index
newdf.set_index('Order Date', inplace=True)
#step2: aggregate sales by month to  see the monthly (ME=bulan) sales trend
monthly_sales = newdf['Sales'].resample('ME').sum()

#step3:plot the monthly sales trend using matplotlib and seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Monthly Sales Trend - Sample Superstore')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

#step4 : Optional - Further analysis (yearly, quarterly trends)
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)
yearly_sales = df['Sales'].resample('Y').sum()
plt.figure(figsize=(10,5))
sns.barplot(x=yearly_sales.index.year, y=yearly_sales.values, palette='viridis')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```
